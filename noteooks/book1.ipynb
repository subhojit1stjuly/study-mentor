```python
def is_palindrome(s: str) -> bool:
    right = len(s) - 1
    left = 0
    while left < right:
        while not s[left].isalnum():
            left += 1
        while not s[right].isalnum():
            right -= 1
        if s[left].lower() != s[right].lower():
            return False
        else:
            left += 1
            right -= 1
    return True
```

```python
def max_profit(prices: list[int]) -> int:
    left = 0
    right = 1
    profit = 0
    while left < right and right < len(prices) :
        current_profit = prices[right] - prices[left]
        if current_profit > profit :
            profit = current_profit
        if prices[left] > prices[right]:
            left = right
        right += 1        
    return profit
```

```python
def max_profit(prices: list[int]) -> int:
   left = 0
   profit = 0
   if len(prices) < 1:
      return profit
   for i in range(1,len(prices)):
      profit = max(profit,(prices[i]-prices[left]))
      if prices[i] < prices[left]:
         left = i
   return profit 
max_profit([7, 1, 5, 3, 6, 4])
```

```python
def length_of_longest_substring(s: str) -> int:
   left = 0
   window_size = 0
   seen = {}
   for right in range(len(s)):      # start from 0, not 1
      while s[right] in seen:
         seen.pop(s[left])
         left += 1
      seen[s[right]] = 1  
      window_size = max(window_size, (right - left + 1))
    
   return window_size
length_of_longest_substring("dvdf")
```

```python
def max_area(height: list[int]) -> int:
    area,left,right = 0,0,len(height)-1
    while left < right:
        area = max(area,min(height[left], height[right]) * (right - left))
        if height[left] < height[right]:
            left += 1
        else:
            right -=1

    return area
max_area([1, 8, 6, 2, 5, 4, 8, 3, 7])
```

```python
def product_except_self(nums: list[int]) -> list[int]:
    answer = [1] * len(nums)
    running = 1
    for i in range(len(nums)):
        answer[i] = running
        running *= nums[i]
    running =  1
    for i in range(len(nums)-1,-1,-1):
        answer[i] *= running
        running *= nums[i]
    return answer
product_except_self([1, 2, 3, 4])
```

```python
def max_subarray(nums: list[int]) -> int:
    running_sum = 0
    best = nums[0]
    for i in range(len(nums)):
        running_sum = max(nums[i], running_sum + nums[i])
        best = max(best,running_sum)
    return best
max_subarray([-3,-1,-2])
```
