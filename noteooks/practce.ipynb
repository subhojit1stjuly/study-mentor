```python
def two_sum_sorted(numbers: list[int], target: int) -> list[int]:
    seen = {}
    for i in range(len(numbers)):
        deducted_value = target - numbers[i]
        if deducted_value in seen:
            return [seen[deducted_value]+1,i+1]
        seen[numbers[i]] = i
    return []
two_sum_sorted([-1,0], -1) 
```

```python
def char_replacement(s: str, k: int) -> int:
    left,window_size,right = 0,0,0
    seen = {}
    for char in s:
        seen[char] = seen.get(char,0) + 1
    while right < len(s):
        if (window_size - seen[s[right]]) <= k:
            window_size = max(window_size,(right - left + 1))
            right += 1
        else:
            left += 1
        
    return window_size
char_replacement("ABABBA", 2)
```

```python
def move_zeroes(nums: list[int]) -> list:
    left = 0
    for right in range(1,len(nums)):
        if nums[left] == 0 and nums[right] != 0:
            nums[left] += nums[right]
            nums[right] = nums[left] - nums[right]
            nums[left] = nums[left] - nums[right]
        # check if window is invalid-valid
        if nums[left] != 0 :
            left += 1
    return nums
move_zeroes([0,1,0,3,12])
```

```python
def remove_duplicates(nums: list[int]) -> int:
    index,uniques = 1,0
    while index < (len(nums) - 1):
        while nums[index] == (nums[index + 1]):
            index += 1
        uniques += 1
        index += 1
    return uniques
remove_duplicates([0,0,1,1,1,2,2,3])
```

```python
def search_insert(nums: list[int], target: int) -> int:
    for i in range(len(nums)):
        if nums[i]>=target:
            return i
    return len(nums)
search_insert([1,3,5,6], 7)
```

```python
def search_insert(nums: list[int], target: int) -> int:
    left,right = 0, len(nums) - 1
    while left<right:
        mid = (right + left) // 2  + 1
        if nums[mid] == target:
            return mid
        elif nums[mid] > target:
            right = mid - 1
        else:
            left = mid + 1
    return left
search_insert([1,3,5,6], 7)
```
